**Imports**

```python
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.metrics import r2_score
import sys
sys.path.append('../')
from services.db_postgres import run_query
```

**Load the data**

```python
sql='''SELECT *
FROM world_happiness
'''
happiness_df = run_query(sql)
```

```python
happiness_df.head(3)
```

**R-squared**

```python
r2 = r2_score(happiness_df['happiness_score'], happiness_df['happiness_prediction'])
print(f'R-squared (R2) Score: {r2}')
```

```python
sns.set_theme(style="darkgrid")
plt.figure(figsize=(10, 6))
sns.lineplot(data=happiness_df, x='id', y='happiness_score', label='Observed')
sns.lineplot(data=happiness_df, x='id', y='happiness_prediction', label='Estimated')
plt.ylabel('Happiness Score')
plt.title('Observed and Estimated Happiness Scores')
plt.legend()
plt.show()
```
